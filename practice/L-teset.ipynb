# Lambda와 map, filter, reduce
## 1. Lambda

```python
# 함수 사용과 기초적인 lambda 사용법

## 함수 사용시
def double(x):
    return 2*x

list(map(double,[1,2,3]))

## lambda 사용시
(lambda x : x*2)(8)

### 람다에 조건 걸기
a=[4,5,6,7,8,9]
map((lambda x : x/2 if x%2==0 else x),a)

del a
```

## 2. map 사용하기

```python
a=[1,2,3,4,5]
b=[5,4,3,2,1]

c=list(map((lambda x,y:x*y),a,b))
```

## 3. filter 사용하기

```python
# lambda로 필터링 해보기

# 1. function을 통해서
test = [4,6,8,10,12,14]

def f(x):
    if 5<x<10 :
        return x
    else :
        return

list(map(f,test))    
```

```python
# 2.function을 간단히
def g(x):
    return 5<x<10

list(filter(g,test))

```

```python
# 3. lambda를 이용해서!
list(filter((lambda x: x if 5<x<10 else None),test))

```

```python
# 4.lambda 식을 더 간단하게
list(filter((lambda x: 5<x<10),test))
```

## 4.reduce 사용하기  
reduce는 반복 가능한 객체의 각 요소를 지정된 함수로 처리 후 누적 반환  
(python 3부터는 내장이 아니므로 import 필요)

```python
# 함수 불러오고 변수 설정
from functools import reduce

a=[1,2,3,4,5]
```

```python
## 1. 함수를 정의 하여 reduce
def f(x,y):
    return x+y

def g(x,y):
    return x-y

def h(x,y):
    return x*y #팩토리얼과 같아짐!

print(reduce(f,a))
print(reduce(g,a))
print(reduce(h,a))
```

```python
## 2. lambda 이용하기

print(reduce(lambda x,y : x+y , a))
print(reduce(lambda x,y : x-y , a))
print(reduce(lambda x,y : x*y , a))
```
