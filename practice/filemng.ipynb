```python
with open ('hello.txt','r') as file:
    txt=file.read()
    print(txt)
```

```python
with open('hello.txt','w') as file:
    for i in range(3):
        file.write("Hi everybody! {i}")
```

```python
with open('hello.txt','w') as file:
    for i in range(3):
        file.write(f'Hi everybody! {i}\n')
```
